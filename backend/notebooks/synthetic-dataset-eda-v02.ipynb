```python
import seaborn as sns

# Analyze the distribution of estimated property prices
if 'estimated_price' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['estimated_price'], bins=30, kde=True)
    plt.title("Estimated Price Distribution")
    plt.xlabel("Estimated Price")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Analyze the distribution of land size and floor area
if 'land_size_num' in df.columns and 'floor_area_num' in df.columns:
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    sns.histplot(df['land_size_num'], bins=30, kde=True)
    plt.title("Land Size Distribution")
    plt.xlabel("Land Size")
    
    plt.subplot(1, 2, 2)
    sns.histplot(df['floor_area_num'], bins=30, kde=True)
    plt.title("Floor Area Distribution")
    plt.xlabel("Floor Area")
    
    plt.tight_layout()
    plt.show()
```

```python
# Analyze the distribution of properties by suburb and state
if 'suburb' in df.columns:
    print("Top 10 Suburbs by Property Count:")
    print(df['suburb'].value_counts().head(10))

if 'state' in df.columns:
    print("\nProperties by State:")
    print(df['state'].value_counts())
```

```python
# Analyze the distribution of properties by year built
if 'year_built' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['year_built'].dropna(), bins=30, kde=True)
    plt.title("Distribution of Year Built")
    plt.xlabel("Year Built")
    plt.ylabel("Frequency")
    plt.show()
```
